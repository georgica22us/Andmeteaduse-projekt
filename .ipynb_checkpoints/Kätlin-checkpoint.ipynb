```python
#Kätlin's notebook
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("red_wine.csv")
# andmed: chlorides, free sulf, total sulf, density, sulphates
```

```python
data
```

```python
def jagunemine(a):
    Halb = 0
    Keskmine = 0
    Hea = 0
    Super = 0
    Tulemus = []
    for i in range (len(a)):
        if a[i] == 3 or a[i] == 4:
            Halb += 1
        elif a[i] == 5 or a[i] == 6:
            Keskmine += 1
        elif a[i] == 7:
            Hea +=1
        else:
            Super +=1
    Halb_prc = Halb/len(a)
    Keskmine_prc = Keskmine/len(a)
    Hea_prc = Hea/len(a)
    Super_prc = Super/len(a)
    Tulemus = [Halb_prc, Keskmine_prc, Hea_prc, Super_prc]
    return Tulemus
```

```python
def barplot(X, Y, Title):
    plt.figure(figsize = (5, 5))
    plt.bar(X, Y)
    plt.title(Title)
    plt.show()
```

```python
def keskmine(a):
    summa = 0
    for i in range(len(a)):
        summa += a[i]
    return summa/len(a)
```

```python
abi = ["Bad wine", "Average wine", "Good wine", "Superb wine"]
```

```python
#Chlorides agains quality:-----------------------------------------------------------------------------------------------------
```

```python
x = data["chlorides"]
y = data["quality"]
sns.scatterplot(x,y)
```

```python
plt.bar(data["quality"], data["chlorides"])
plt.xlabel("Quality")
plt.ylabel("Chlorides")
```

```python
data["chlorides"].sort_values()
```

```python
x = data["chlorides"]
y = data["quality"]
f1=[]
f2=[]
f3=[]
f4=[]
f5=[]
f6=[]

for i in range(len(x)):
    add = y[i]
    if x[i] <= 0.06:
        f1.append(add)
    elif x[i] > 0.06 and x[i] <= 0.07:
        f2.append(add)
    elif x[i] > 0.07 and x[i] <= 0.075:
        f3.append(add)
    elif x[i] > 0.075 and x[i] <= 0.08:
        f4.append(add)
    elif x[i] > 0.08 and x[i] <= 0.09:
        f5.append(add)
    else:
        f6.append(add)
        
print(len(f1))
print(len(f2))
print(len(f3))
print(len(f4))
print(len(f5))
print(len(f6))
```

```python
kesk1 = keskmine(f1)
kesk2 = keskmine(f2)
kesk3 = keskmine(f3)
kesk4 = keskmine(f4)
kesk5 = keskmine(f5)
kesk6 = keskmine(f6)

x_v = ["0-0.06", "0.06-0.07","0.07-0.075", "0.075-0.08", "0.08-0.09", "0.09+"]
y_kesk = [kesk1, kesk2, kesk3, kesk4, kesk5, kesk6]
plt.bar(x_v, y_kesk)
plt.xlabel("Chlorides")
plt.ylabel("Average wine quality")
plt.title("Chlorides impact on quality")
plt.show()
```

```python
fixed_1 = jagunemine(f1)
fixed_2 = jagunemine(f2)
fixed_3 = jagunemine(f3)
fixed_4 = jagunemine(f4)
fixed_5 = jagunemine(f5)
fixed_6 = jagunemine(f6)

barplot(abi, fixed_1, "Analysis of wines with chlorides below 0.06")
barplot(abi, fixed_2, "Analysis of wines with chlorides of 0.06-0.07")
barplot(abi, fixed_3, "Analysis of wines with chlorides of 0.07-0.075")
barplot(abi, fixed_4, "Analysis of wines with chlorides of 0.075-0.08")
barplot(abi, fixed_5, "Analysis of wines with chlorides of 0.08-0.09")
barplot(abi, fixed_6, "Analysis of wines with chlorides over 0.09")
```
